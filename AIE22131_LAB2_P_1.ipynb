```python
#Sum of first N natural numbers using iterative and recursive algorithms. 
import time
import matplotlib.pyplot as plt

def iterative_sum(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum

def recursive_sum(n):
    if n == 0:
        return 0
    else:
        return n + recursive_sum(n-1)

def measure_time(func, n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

# Varying N's value
N_values = list(range(1, 1001, 10))

iterative_times = []
recursive_times = []

for n in N_values:
    iterative_times.append(measure_time(iterative_sum, n))
    recursive_times.append(measure_time(recursive_sum, n))

# Plotting the results
plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time taken to calculate sum of first N natural numbers')
plt.legend()
plt.show()
```

```python
#Perform Linear and Binary searches for array of 10000 elements
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generate random array of 10,000 elements
arr = [random.randint(1, 1000) for _ in range(10000)]

# Perform searches and measure time
search_keys = [random.randint(1, 1000) for _ in range(5)]

linear_search_times = []
binary_search_times = []

for key in search_keys:
    start_time = time.time()
    linear_search(arr, key)
    end_time = time.time()
    linear_search_times.append(end_time - start_time)

    start_time = time.time()
    binary_search(arr, key)
    end_time = time.time()
    binary_search_times.append(end_time - start_time)

# Plotting the results
plt.plot(search_keys, linear_search_times, label='Linear Search')
plt.plot(search_keys, binary_search_times, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Time (seconds)')
plt.title('Time taken for searching for 5 different keys')
plt.legend()
plt.show()
```

```python
#to convert string of digits to integers it represents 
def string_to_integer(s):
    # Base case: if the string is empty, return 0
    if not s:
        return 0
    
    # Convert the first character to an integer
    digit = int(s[0])
    
    # Recursive call to process the remaining characters
    remaining_string = s[1:]
    remaining_integer = string_to_integer(remaining_string)
    
    # Combine the current digit with the integer obtained from the remaining characters
    result = digit * (10 ** len(remaining_string)) + remaining_integer
    
    return result

# Test the function
input_string = input("Enter a string of digits: ")
result_integer = string_to_integer(input_string)
print("The integer representation is:", result_integer)
```

```python
#Code to reverse the string
def reverse_string(s):
    # Base case: if the string is empty or has only one character, return the string
    if len(s) <= 1:
        return s
    # Recursive case: return the last character of the string concatenated with the reverse of the rest of the string
    else:
        return s[-1] + reverse_string(s[:-1])

# Test the function
input_string = "pots@pans"
result_string = reverse_string(input_string)
print("The reverse of", input_string, "is:", result_string)
```

```python
#To check if string is palindrome or not
def is_palindrome(s):
    # Base case: if the string is empty or has only one character, it is a palindrome
    if len(s) <= 1:
        return True
    # Recursive case: check if the first and last characters are the same,
    # then recursively check if the substring between them is also a palindrome
    else:
        return s[0] == s[-1] and is_palindrome(s[1:-1])

# Test the function
input_string = "racecar"
if is_palindrome(input_string):
    print(input_string, "is a palindrome")
else:
    print(input_string, "is not a palindrome")

input_string = "gohangasalamiimalasagnahog"
if is_palindrome(input_string):
    print(input_string, "is a palindrome")
else:
    print(input_string, "is not a palindrome")
```
